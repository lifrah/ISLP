```python
import numpy as np
Auto = pd.read_csv("../data sets/Auto.csv", na_values=['?'])
Auto = Auto.dropna()
Auto.columns
```

### (a)

```python
Auto.head()
# Quantitative: displacement, horsepower, weight. 
# Qualitative: mpg, cylinders, year, name, origin. 
```

### (b)

```python
disp, hp, wei = Auto[['displacement']], Auto[['horsepower']], Auto[['weight']]
print("min disp, hp, wei:", np.min(disp), np.min(hp), np.min(wei),"\nmax disp, hp, wei:", np.max(disp), np.max(hp), np.max(wei))
```

### (c)

```python
print("mean disp, hp, wei:", np.mean(disp), np.mean(hp), np.mean(wei),"\nstd disp, hp, wei:", np.std(disp), np.std(hp), np.std(wei))
```

### (d)

```python
keep_rows = np.ones(Auto.shape[0], bool)

for i in range(10,86):
    keep_rows[i] = False

Auto_re = Auto.iloc[keep_rows]
disp, hp, wei = Auto_re[['displacement']], Auto_re[['horsepower']], Auto_re[['weight']]
print("mean disp, hp, wei:", np.mean(disp), np.mean(hp), np.mean(wei),"\nstd disp, hp, wei:", np.std(disp), np.std(hp), np.std(wei))
```
